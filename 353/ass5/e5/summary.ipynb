```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data-6.csv')
```

```python
max_x = df['x'].max()
print('max_x:', max_x)
max_y = df['y'].max()
print('max_y:',max_y)

min_x = df['x'].min()
print('min_x:',min_x)
min_y = df['y'].min()
print('min_y:',min_y)

std_x = df['x'].std()
print('std_x:', std_x)
std_y = df['y'].std()
print('std_y:',std_y)

mean_x = df['x'].mean()
print('mean_x:',mean_x)
mean_y = df['y'].mean()
print('mean_y:',mean_y)

corr = df['x'].corr(df['y'])
print('corr:', corr)
```

```python
plt.plot(df['x'].values, df['y'].values, 'b.', alpha=0.5)
plt.show
```
